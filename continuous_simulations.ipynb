### 2. Queueing System: compare analytical/empirical results

```python
system_params = [
    {
        "R" : 2,
        "diffusion_params": [
            {"drift" : 3, "sigma": 2},
            {"drift" : -2, "sigma": 2}
        ],
        "diffusion_borders": [
            (0, 40), (20, 60)
        ],
        "level_costs": [2, 4],
        "transition_costs": [10]
    },
    {
        "R" : 3,
        "diffusion_params": [
            {"drift" : 0, "sigma": 10},
            {"drift" : 0, "sigma": 10},
            {"drift" : 0, "sigma": 10}
        ],
        "diffusion_borders": [
            (0, 40), (20, 60), (50, 80)
        ],
        "level_costs": [2, 3, 7],
        "transition_costs": [10, 20]
    },
    {
        "R" : 5,
        "diffusion_params": [
            {"drift" : 2.5, "sigma": 8},
            {"drift" : 1, "sigma": 8},
            {"drift" : 0, "sigma": 8},
            {"drift" : -1.5, "sigma": 8},
            {"drift" : -3, "sigma": 8}
        ],
        "diffusion_borders": [
            (0, 30), (20, 50), (40, 65), (60, 90), (70, 100)
        ],
        "level_costs": [1, 2, 3, 4, 5],
        "transition_costs": [10, 20, 30, 40]
    }
]

def test_queueing_system(system_params, T = 10000, N = 1000):
    qsystem = QueueingSystem(**system_params)
    start = time.time()
    results = Parallel(n_jobs=-1)( # , backend="multiprocessing"
        delayed(qsystem.empirical_cost_estimation)(T) 
        for _ in tqdm(range(N), desc="Running Simulations")
    )
    emp_eval = np.round(np.mean(results), 5)
    emp_std = np.round(np.std(results, ddof=1), 5)
    emp_end_eval = time.time() - start
    print(f"Simulations     - time: {np.round(emp_end_eval, 5)} sec.; avg. unit cost: {emp_eval} in [{emp_eval - 1.96 * emp_std / np.sqrt(N)}, {emp_eval + 1.96 * emp_std / np.sqrt(N)}]")
    start = time.time()
    opt_eval = qsystem.analytical_cost_estimation()
    opt_end_eval = time.time() - start
    print(f"Optimal         - time: {np.round(opt_end_eval, 5)} sec.; avg. unit cost: ", np.round(opt_eval, 5))
```

```python
print("R = 2, N = 60: ")
test_queueing_system(system_params[0], T = 1000000, N = 10)
print()
print("R = 3, N = 80: ")
test_queueing_system(system_params[1], T = 1000000, N = 10)
print()
print("R = 5, N = 100: ")
test_queueing_system(system_params[2], T = 10000000, N = 10)
```
